```python
import pandas as pd
```

```python
pe = pd.read_csv('UK PEBand - UKPE.csv')
cum = pd.read_csv('UK PEBand - backtestReturn.csv')
to_csv = "UKmean.csv"
```

```python
pe.date = pd.to_datetime(pe.date)
```

```python
cum.RnY = pd.to_datetime(cum.RnY)
```

```python
print(pe['date'].dt.tz)
print(cum['RnY'].dt.tz)
```

```python
pe.date = pe.date.dt.tz_localize('UTC')
```

```python
cum['date'] = cum.RnY
```

```python
pe.set_index('date', inplace=True)
cum.set_index('date', inplace=True)
```

```python
merge = pe.merge(cum, left_index=True, right_index=True)
```

```python
merge.columns
```

```python
merge.to_csv(to_csv)
```
